```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv('sales_df.csv')
sales_data.head()
```

Total Sales

```python
sales_data['date_ordered'] = pd.to_datetime(sales_data['date_ordered'])

# Group by date and sum the quantities
total_sales = sales_data.groupby('date_ordered')['quantity'].sum()

plt.figure(figsize=(12, 6))
total_sales.plot(kind='line', color='skyblue')  # Continuous line plot for daily sales
plt.title('Sales Distribution Throughout the Year')
plt.xlabel('Date')
plt.ylabel('Total Sales Quantity')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.legend()
plt.tight_layout()
plt.show()

```

Yearly Sales

```python

sales_data['date_ordered'] = pd.to_datetime(sales_data['date_ordered'])


yearly_sales = sales_data[sales_data['date_ordered'].dt.year == 2023].groupby('date_ordered')['quantity'].sum()

plt.figure(figsize=(12, 6))
yearly_sales.plot(kind='line', color='skyblue')  # Continuous line plot for daily sales
plt.title('Sales Distribution Throughout the Year')
plt.xlabel('Date')
plt.ylabel('Total Sales Quantity')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.legend()
plt.tight_layout()
plt.show()
```

Google trends

```python
trends = pd.read_csv('./datasets/Google_trends_data/Bicycle_search.csv')

trends['Week'] = pd.to_datetime(trends['Week'])


yearly_trends = trends[trends['Week'].dt.year == 2023].groupby('Week')['Bicycle'].sum()

# Assuming 'trends' DataFrame has 'week' and 'bicycle' columns
plt.figure(figsize=(12, 6))
plt.plot(trends['Week'], trends['Bicycle'], linestyle='-', color='skyblue')
plt.title('Bicycle Search Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Bicycle Search Volume')
plt.show()
trends.head()
```
