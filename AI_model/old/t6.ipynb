```python
from datetime import datetime


# Step 3: Data Preprocessing
data['date_ordered'] = pd.to_datetime(data['date_ordered'])
data = data.sort_values(['bike_name', 'date_ordered'])  # Sort by product and date

# Step 4: Split Data into Training and Test Set
# Train on data before 2023
train_data = data[data['date_ordered'] < '2023-01-01']
test_data = data[data['date_ordered'] >= '2023-01-01']

# Feature Engineering on Training Data
train_data['year'] = train_data['date_ordered'].dt.year
train_data['month'] = train_data['date_ordered'].dt.month
train_data['day'] = train_data['date_ordered'].dt.day
train_data['day_of_week'] = train_data['date_ordered'].dt.dayofweek

# Add seasonality features (cosine/sine transformation)
train_data['month_sin'] = np.sin(2 * np.pi * train_data['month'] / 12)
train_data['month_cos'] = np.cos(2 * np.pi * train_data['month'] / 12)

#
# Special events for sales spikes
christmas = [datetime(year, 12, 25) for year in range(2021, 2024)]
black_friday = [datetime(year, 11, 26) for year in range(2021, 2024)]
tour_de_france = [datetime(year, 7, 1) for year in range(2021, 2024)]
summer_season = [(datetime(year, 6, 1), datetime(year, 8, 31)) for year in range(2021, 2024)]

# Now proceed with the feature engineering
train_data['is_christmas'] = train_data['date_ordered'].apply(lambda x: 1 if x in christmas else 0)
train_data['is_black_friday'] = train_data['date_ordered'].apply(lambda x: 1 if x in black_friday else 0)
train_data['is_tour_de_france'] = train_data['date_ordered'].apply(lambda x: 1 if any(abs((x - event).days) <= 30 for event in tour_de_france) else 0)
train_data['is_summer'] = train_data['date_ordered'].apply(lambda x: 1 if any(start <= x <= end for start, end in summer_season) else 0)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
# Include bike type as a categorical feature
train_data['bike_type_encoded'] = le.fit_transform(train_data['bike_type'])

# Lag features for each product's previous quantity
train_data['quantity_lag1'] = train_data.groupby('bike_name')['quantity'].shift(1)
train_data['rolling_avg_7'] = train_data.groupby('bike_name')['quantity'].rolling(window=7).mean().reset_index(0, drop=True)

# Drop rows with NaN values from lag/rolling features
train_data = train_data.dropna()

# Feature Engineering on Test Data (2023)
test_data['year'] = test_data['date_ordered'].dt.year
test_data['month'] = test_data['date_ordered'].dt.month
test_data['day'] = test_data['date_ordered'].dt.day
test_data['day_of_week'] = test_data['date_ordered'].dt.dayofweek

# Add seasonality features (cosine/sine transformation)
test_data['month_sin'] = np.sin(2 * np.pi * test_data['month'] / 12)
test_data['month_cos'] = np.cos(2 * np.pi * test_data['month'] / 12)

# Event-based features
test_data['is_christmas'] = test_data['date_ordered'].apply(lambda x: 1 if x in christmas else 0)
test_data['is_black_friday'] = test_data['date_ordered'].apply(lambda x: 1 if x in black_friday else 0)
test_data['is_tour_de_france'] = test_data['date_ordered'].apply(lambda x: 1 if any(abs((x - event).days) <= 30 for event in tour_de_france) else 0)
test_data['is_summer'] = test_data['date_ordered'].apply(lambda x: 1 if any(start <= x <= end for start, end in summer_season) else 0)

# Include bike type as a categorical feature
test_data['bike_type_encoded'] = le.transform(test_data['bike_type'])

# Lag features for the test data
test_data['quantity_lag1'] = test_data.groupby('bike_name')['quantity'].shift(1)
test_data['rolling_avg_7'] = test_data.groupby('bike_name')['quantity'].rolling(window=7).mean().reset_index(0, drop=True)

# Drop rows with NaN values from lag/rolling features
test_data = test_data.dropna()

# Step 5: Encode Categorical Data (Product)

train_data['product_encoded'] = le.fit_transform(train_data['bike_name'])
test_data['product_encoded'] = le.transform(test_data['bike_name'])

# Step 6: Define Features and Target (including event-based features)
X_train = train_data[['product_encoded', 'bike_type_encoded', 'year', 'month', 'day', 'day_of_week', 'month_sin', 'month_cos', 
                      'quantity_lag1', 'rolling_avg_7', 'is_christmas', 'is_black_friday', 'is_tour_de_france', 'is_summer']]
y_train = train_data['quantity']

X_test = test_data[['product_encoded', 'bike_type_encoded', 'year', 'month', 'day', 'day_of_week', 'month_sin', 'month_cos', 
                    'quantity_lag1', 'rolling_avg_7', 'is_christmas', 'is_black_friday', 'is_tour_de_france', 'is_summer']]
y_test = test_data['quantity']

print(X_train.head())

```

```python
# Calculate MAPE if desired
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
